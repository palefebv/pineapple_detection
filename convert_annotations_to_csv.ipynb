```python
import xml.etree.ElementTree as ET
import pandas as pd
import os, os.path
```

```python
cols = ['filename','width','height','class','xmin','ymin','xmax','ymax','source']
df = pd.DataFrame(columns=cols)  # Write in DF and transpose it
print(df)
```

```python
annotations = [name for name in os.listdir('dataset/annotations')]
```

```python
# xml_data = open('dataset/annotations/pineapple13.xml', 'r').read()  # Read file
# root = ET.XML(xml_data)  # Parse XML
```

```python
for annotation in annotations:
    xml_data = open('dataset/annotations/'+ annotation, 'r').read()  # Read file
    root = ET.XML(xml_data)  # Parse XML
    for i, child in enumerate(root):
        if child.tag == 'filename':
            filename = child.text
        if child.tag == 'path':
            source = child.text
        if child.tag == 'size':
            for subchild in child:
                if subchild.tag == 'width':
                    width = subchild.text
                if subchild.tag == 'height':
                    height = subchild.text
        if child.tag == 'object':
            for subchild in child:
                if subchild.tag == 'name':
                    name = subchild.text
                if subchild.tag == 'bndbox':                
                    for subsubchild in subchild:
                        if subsubchild.tag == 'xmin':
                            xmin = subsubchild.text
                        if subsubchild.tag == 'ymin':
                            ymin = subsubchild.text
                        if subsubchild.tag == 'xmax':
                            xmax = subsubchild.text
                        if subsubchild.tag == 'ymax':
                            ymax = subsubchild.text
                    new_row = {'filename':filename,'width':width,'height':height,'class':name,'xmin':xmin,'ymin':ymin,'xmax':xmax,'ymax':ymax,'source':source}
                    df = df.append(new_row, ignore_index=True)
```

```python
df
```

```python
train=df.sample(frac=0.8,random_state=42) #random state is a seed value
test=df.drop(train.index)
```

```python
train.to_csv (r'train_labels.csv', index = False, header=True)
test.to_csv (r'test_labels.csv', index = False, header=True)
```
